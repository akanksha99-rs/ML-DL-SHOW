A notebook to explain how Multivariate Linear Regression works

## Importing necessary libraries

* pandas
* matplotlib
    * pyplot
* seaborn
* numpy

```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("50_Startups_csv.csv")
dataset.fillna(0)
dataset.tail()
```

```python
dataset.corr()
```

```python
'''Scatter plot of R&D vs Profit and Marketing vs Profit'''
plt.scatter( dataset["R&D Spend"], dataset["Profit"],color="rgb" )
plt.xlabel("R&D Spend -------->")
plt.ylabel("Profit --------->")
plt.title("R&D vs Profit")
plt.show()

plt.scatter( dataset["Marketing Spend"], dataset["Profit"], color="rgb" )
plt.xlabel("Marketing Spend -------->")
plt.ylabel("Profit --------->")
plt.title("Marketing vs Profit")
plt.show()
```

## Building model

```python
X = dataset.iloc[ :, [0,2] ].values
Y = dataset.iloc[ :,[-1]].values
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/10)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
from sklearn.metrics import r2_score
pred = regressor.predict(X_test)
print(pred)
```

```python
acc = r2_score(Y_test, pred)
acc
```
